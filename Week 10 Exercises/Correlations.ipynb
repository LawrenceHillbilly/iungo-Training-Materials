```python
df = pd.read_csv('Correlations.csv', encoding= 'utf-8')

correlation_coefficient, p_value = stats.pearsonr(df['Machine Age'], df['Maintenance Cost'])

plt.figure(figsize=(10, 6))
plt.scatter(df['Machine Age'], df['Maintenance Cost'])

plt.xlabel('Machine Age / Years')
plt.ylabel("Maintenance Cost / UKP")
plt.title(f"Correlation between Machine Age and Maintenance Cost \n Correlation Coefficient = {np.round(correlation_coefficient,3)}")
plt.xticks(df['Machine Age'])
plt.show()

#Defect Rate,Production Speed
```

```python
df = pd.read_csv('Correlations.csv', encoding= 'utf-8')

correlation_coefficient, p_value = stats.pearsonr(df['Defect Rate'], df['Operator Experience (Years)'])

plt.figure(figsize=(10, 6))
plt.scatter(df['Defect Rate'], df['Operator Experience (Years)'])

plt.xlabel('Operator Experience (Years)')
plt.ylabel("Defect Rate")
plt.title(f"Correlation between Defect Rate and Operator Experience (Years) \n Correlation Coefficient = {np.round(correlation_coefficient,3)}")
plt.xticks(df['Operator Experience (Years)'])
plt.show()
```
