```python
#import the maths libraries 
import numpy as np
from matplotlib import pyplot as plt

#Define the possible values
LayerOptions = [1,2,3,4,5]

#Define the number of selections (100,000 - 1 Million). The higher the number, the more certain you can be
num_selections = 1000000
selected_layers = np.random.choice(LayerOptions, num_selections) 

#Count the occurrences of each possible layer
values, counts = np.unique(selected_layers, return_counts=True)

#Calculate the probabilities (counts / number of selections)
probabilities = counts / num_selections

print("Probabilities for each layer option:", probabilities)

#Plot the bar graph
plt.bar(values, probabilities, color = '#00BEEA', edgecolor = 'black')
plt.xlabel("Number of Layers")
plt.ylabel("Probability")
plt.title("Discrete Uniform Distribution: Random Hyperparameter Selection over 1 Milion tests")
plt.xticks(values)
plt.show()
```
