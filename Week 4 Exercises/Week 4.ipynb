```python
#Create a function called get grade. It should take one parameter called score. Assume for now that score, 
# is a number between 1 and 100. Using grade boundaries of your choosing, return a grade string based on the score

def get_grade(score):
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    elif score >= 50:
        grade = "E"
    else:
        grade = "F"
    
    return grade
```

```python
#Create two dictionaries called student_1 and student_2. Each ditionary should contain keys
#"student_id" (number), "student_name" (string) and "exam_score" (number between 1 and 100)
#put the dictionaries into a list called student data

student_1 = {
    "student_id" : 1,
    "student_name" : "Jane Smith",
    "exam_score" : 92
}

student_2 = {
    "student_id" : 2,
    "student_name" : "John Jones",
    "exam_score" : 74
}

student_3 = {
    "student_id" : 58008,
    "student_name" : "Hugh Jarse",
    "exam_score" : 85
}

student_4 = {
    "student_id" : 1337,
    "student_name" : "Melon Husk",
    "exam_score" : 62
}

student_data = [student_1, student_2, student_3, student_4]

#for every student in student_data
for student in student_data:
    # add a grade key using the get grade function you wrote earlier
    student["grade"] = get_grade(student["exam_score"])

for student in student_data:
    print(student)
```

```python
def get_discount(base_price):
    return base_price * 0.8

products = [
    {"name" : "TV", "price" : 500},
    {"name" : "Console", "price" : 700}

]

for product in products:
    #Create a new field in each product with the vakue being the discounted price
    product["discount_price"] = get_discount(product["price"])

print(products)
```

```python
import json
#reading json files
# 
# #Read the student json data into python. Using a for loop, add a grade field to each of the
#1000 students. Load the transformed data to a file called students_transformed.json
# 
with open("255003_student_data.json", "r+") as infile:
    student_data=json.load(infile)

#for every student in student_data
for student in student_data:
    # add a grade key using the get grade function you wrote earlier
    student["grade"] = get_grade(student["exam_score"])

with open("students_transformed.json", "w") as outfile:
    json.dump(student_data, outfile, indent=4)

for student in student_data:
    print(student)
```
