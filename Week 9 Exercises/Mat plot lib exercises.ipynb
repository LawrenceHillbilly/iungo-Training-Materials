```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
emloyees = {
    "Name" : ["Matt", "Dan", "Ploy", "Ruben", "Arif", "Jade"],
    "Delivery Hours" : [16, 30, 2, 32, 16,5]
}
```

```python
df = pd.DataFrame(emloyees)
df
```

```python
plt.figure(figsize=(10,6))
plt.title("Delivery Hours for iungo Employees")
plt.xlabel("Tutor")
plt.ylabel("Delivery Hours")
plt.bar(df["Name"], df["Delivery Hours"],color=['red','blue','green','cyan','magenta','yellow'],width=0.8,)
plt.show
```

```python
plt.figure(figsize=(8,4))
plt.title("Delivery Hours Linegraph")
plt.xlabel("Tutor")
plt.ylabel("Delivery Hours")
plt.plot(df["Name"], df["Delivery Hours"], marker = "X", linestyle = "--")
plt.show
```

```python

df2 = pd.read_csv("sales.csv")

df2["Percentage Change"] = df2['Sales'].pct_change() * 100
df2["Percentage of Recalls"] = 100*(df2['Sales']-df2['Recalls'])/df2['Sales']

plt.figure(figsize=(8,6))
plt.title("Sales per Annum")
plt.xlabel("Year")
plt.ylabel("Annual Sales")
plt.plot(df2["Years"], df2["Sales"], marker = "D")
plt.plot(df2["Years"], df2["Recalls"], marker = "o",color="Red", linestyle = ":")

plt.show

plt.figure(figsize=(8,6))
plt.title("Percentage Increase over time")
plt.xlabel("Year")
plt.grid(True)
plt.plot(df2["Years"], df2["Percentage Change"], marker = "o")
plt.show


plt.figure(figsize=(8,6))
plt.title("Percentage of Recalls per annum")
plt.xlabel("Year")
plt.grid(True)
plt.plot(df2["Years"], df2["Percentage of Recalls"], marker = "o", color = 'green')
plt.show

df2
```
