```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

tosses = np.random.choice(['Heads', 'Tails'], size=1000000)

labels, counts = np.unique(tosses, return_counts=True)

plt.figure(figsize=(6,6))
plt.pie(
    counts, 
    labels=labels,
    startangle=90,
    colors=['#00BEEA', '#9026D8']
)
plt.title("Results of 1,000,000 coin tosses")
plt.axis('equal')
```

```python
#Using np.random.randint, simulate two fair 6-sided dice being thrown 1,000,000 times, and find the sum of both
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

num_rolls = 1000000

die1 = np.random.randint(1, 7, num_rolls)
die2 = np.random.randint(1, 7, num_rolls)

outcomes = die1 + die2


plt.hist(outcomes, bins=np.arange(1.5,13.5,1), edgecolor = 'black', color='#0D8FEE', align='mid')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Probability')
plt.title("Probability Distribution of Rolling Two Dice in 1 Million Simulations")

plt.xticks(range(2,13))
plt.show()
```
