```python
#Using the following dictionary, perform these actions 
products = {
"name" : ["Switch", "TV", "PS5"],
"price" : [200,500,600],
"sales" : [1000, 400, 1500]
}

# Load the dictionary into a pandas dataframe 
df = pd.DataFrame(products)

# Create a new column called revenue  
df["revenue"] = df["price"] * df["sales"] 

# Filter for products with less than 1000 sales 
products_filtered = df.loc[df["sales"] < 1000]

# Add a sales_price column to the filtered dataframe, where each sale price is 30% of the original
products_filtered["sales_price"] = products_filtered["price"] * (1 - 30/100)

products_filtered
```

```python
#group by
product_data = {'Product': ['A', 'B', 'C', 'A', 'B', 'D', 'C', 'E', 'A', 'B', 'B', 'C'],
         'Region': ['North', 'South', 'East', 'North', 'South', 'West', 'East', 'North', 'North', 'South', 'South', 'North'],
         'Sales': [100, 150, 80, 100, 160, 120, 140, 110, 105, 155, 160, 130]}

df=pd.DataFrame(product_data)

#groups by the product and aggregate sales by sum
#df = df.groupby("Product", as_index=False)["Sales"].sum()


#modify to determine the average number of sales accross regions
df = df.groupby("Region", as_index=False)["Sales"].mean()
df
```

```python
#Read employees.json into a dataframe and do the following
df = pd.read_json("employees.json")

#Filter to only include employees from the marketing department
df = df.loc[df["department"] == "Marketing"]

#Filter again for employees in that group that make less than 25000
df = df.loc[df["salary"] < 25000]

#Add a new column called christmas_bonus, where the value if 5% of an employees annual salary 
df["christmas_bonus"] = 0.05 * df["salary"]

df
```

```python
#Apply statement - allows you to write your own functions you can use


def get_raise_eligability(salary):
    if salary < 20000:
        return True
    else:
        return False
    
#it automatically works out which road is the salary to be passed, so don't need to call salary
df["raise_eligable"] = df["salary"].apply(get_raise_eligability)
df
```

```python
#Create a function called get grade. It should take one parameter called score. Assume for now that score, 
# is a number between 1 and 100. Using grade boundaries of your choosing, return a grade string based on the score

def get_grade(score):
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    elif score >= 50:
        grade = "E"
    else:
        grade = "F"
    
    return grade

#Download students.json and load it into a pandas dataframe. 
df = pd.read_json("student_data.json")

# Use the apply function to create a new column in the dataframe called grade. 
df["grade"] = df["exam_score"].apply(get_grade)
df
```
